```python
import yfinance as yf

def get_stock_data(stock, start, end):
    stock = yf.Ticker(stock)
    stock_data = stock.history(start=start, end=end, interval='1d')
    stock_data['Close Returns'] = stock_data['Close'].pct_change()
    return stock_data
```

```python
apple  = get_stock_data('AAPL', '2022-01-01', '2022-12-31')
```

```python
def returns_stringify(df):
    string = ''
    for i in df['Close Returns']:
        if i > 0:
            string += 'U'
        else:
            string += 'D'
    return string

returns_stringify(apple)
```

```python
def generate_ids(df):
    return [int(i > 0) for i in df['Close Returns']]
    
    
generate_ids(apple)
```

```python
def get_stats(ids):
    stats = {}
    for i in zip(ids, ids[1:]):
        stats[i] = stats.get(i, 0) + 1
    return stats

apple_return_stats = get_stats(generate_ids(apple))
apple_return_stats
```

```python
top_pair = max(apple_return_stats, key=apple_return_stats.get)
top_pair
```

```python
def merge(ids, pair, idx):
    newids = []
    i = 0
    while i < len(ids):
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            newids.append(idx)
            i += 2
        else:
            newids.append(ids[i])
            i += 1
    return newids

merged_ids = merge(generate_ids(apple), top_pair, 2)
merged_ids
```

```python
vocab_size = 5
num_merges = vocab_size - 2
ids = generate_ids(apple)

merges = {}
for i in range(num_merges):
    stats = get_stats(ids)
    top_pair = max(stats, key=stats.get)
    idx = 2 + i
    print(f'Merging {top_pair} to {idx}')
    ids = merge(ids, top_pair, idx)
    merges[idx] = top_pair
```

```python
def bpe_train(ids, vocab_size):
    num_merges = vocab_size - 2

    merges = {}
    for i in range(num_merges):
        stats = get_stats(ids)
        top_pair = max(stats, key=stats.get)
        idx = 2 + i
        print(f'Merging {top_pair} to {idx}')
        ids = merge(ids, top_pair, idx)
        merges[idx] = top_pair
        
    return ids, merges

merged_ids, merges = bpe_train(generate_ids(apple), 5)
```

```python
def decode(merges, ids):
    decoded = []
    max_id = max(ids)
    while max_id > 1:
        for id in ids:
            if id == max_id:
                decoded.append(merges[max_id][0])
                decoded.append(merges[max_id][1])
            else:
                decoded.append(id)
        ids = decoded
        decoded = []
        max_id -= 1 
    return ids

decode(merges, merged_ids) == generate_ids(apple)
```

```python
# we want this encode funct
def encode(merges, ids):
    encoded = []
    for idx in sorted(merges.keys()):
        i = 0
        while i < len(ids):
            if i < len(ids) - 1 and (ids[i], ids[i+1]) == merges[idx]:
                encoded.append(idx)
                i += 2
            else:
                encoded.append(ids[i])
                i += 1
        ids = encoded
        encoded = []
    return ids

encode(merges, generate_ids(apple)) == merged_ids
        
```
