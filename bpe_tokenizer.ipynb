```python
import yfinance as yf

def get_stock_data(stock, start, end):
    stock = yf.Ticker(stock)
    stock_data = stock.history(start=start, end=end, interval='1d')
    stock_data['Close Returns'] = stock_data['Close'].pct_change()
    return stock_data
```

```python
get_stock_data('AAPL', '2022-01-01', '2022-12-31')
```
