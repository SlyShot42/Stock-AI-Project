```python
import time
import pandas as pd

def get_sector_and_market_cap(ticker):
    time.sleep(1)  # Add a delay to avoid hitting the API rate limit
    print(ticker)
    ticker_info = yf.Ticker(ticker).info
    return [ticker_info.get('sector'), ticker_info.get('marketCap')]

# Assuming nasdq_NoETF is a DataFrame with ticker symbols as the index
nasdq_NoETF[['Sector', 'Market Cap']] = nasdq_NoETF.index.to_series().apply(get_sector_and_market_cap, result_type='expand')
```
